### [2-2] "/"으로 split해서 정보 확인

* 먼저 어떤 특수문자들이 있는지 파악 필요

```python
users = df.UserInfos.tolist()
```

```python
import re
```

```python
special = re.compile(r'[^ A-Za-z0-9가-힣+]')
result = set()
for user in users:
    result |= set(special.findall(user))
```

```python
json_special = dict()
```

```python
for r in result:
    if r == '/': continue
    json_special[r] = []
    for user in users:
        if r in user: json_special[r].append(user)
```

### [3-1] \n 여러개를 1개로 줄이기

```python
import re

content2 = [re.sub("\n+", "\n", c) for c in content]
```

### [3-2] 본문을 (질문 - 내용)으로 나누기

```python
start_num = []
```

```python
for con in content2:
    if con[0].isdigit(): # 숫자로 시작
        start_num.append(con)
```

```python
len(start_num)
```

```python
# 숫자로 시작하는 경우
# ①, ② : 이런 숫자 모두 1,2,3,로 변경하기
new_contents = []
for con in start_num:
    # [1] \d[. ]?[. ]+(.*?)(\n)+ 사용해서 질문 가져오기
    con = con.replace('①', '1').replace('②', '2').replace('③', '3').replace('④', '4').replace('⑤', '5').replace('⑥', '6')
    con = con.replace('⓵', '1').replace('⓶', '2').replace('⓷', '3')
    qa_dict = dict()
    questions = [q[0] for q in re.findall('\d[. )]?[. )]+(.*?)(\n)+', con)]
    answers = re.sub('\d[. )]?[. )]+(.*?)(\n)+', '###############', con).split("###############")[1:]

    for q, a in zip(questions, answers):
        qa_dict[q] = a.strip()
    new_contents.append(qa_dict)
```

```python
temp = [idx for idx, c in enumerate(new_contents) if not c]
```

```python
len(temp)
```
